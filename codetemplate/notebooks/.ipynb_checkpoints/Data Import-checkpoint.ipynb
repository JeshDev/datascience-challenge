```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Load dataset
cust_lead = pd.read_csv('../../data/CustomerData_LeadGenerator.csv')

# Inspect data
cust_lead.head(10)
```

```python
# Print summary statistics

cust_lead.describe()
```

```python
# Print DataFrame information

cust_lead.info()
```

```python
# Inspect missing values in the dataset

cust_lead.isnull().values.any()
cust_lead.isnull().sum()
```

#### No missing values found

```python
# Checking duplicates

sum(cust_lead.duplicated(subset = 'fakeID')) == 0
```

#### There are no duplicate rows

```python
# Analysing columns with Object datatype and cleaning data

## Removed the row with invalid values and assigned the column a valid data type 
cust_lead = cust_lead[cust_lead.q_OpeningHours != 'ject']
cust_lead["q_OpeningHours"] = cust_lead.q_OpeningHours.astype(float)
```

```python
# Opening Days

print("Opening Days: " + str(sorted(cust_lead['q_OpeningDays'].unique())))
```

```python
# Because it has fixed values, creating a categorical variable from the field
new_col = cust_lead["q_OpeningDays"].astype("category")
cust_lead.insert(loc=12, column='q_OpeningDays_cat', value=new_col)
cust_lead.info()
```

```python
# Analysing dependent variables

kontakt_gewesen = (sum(cust_lead['b_in_kontakt_gewesen'])/len(cust_lead['b_in_kontakt_gewesen'].index))*100
kontakt_gewesen
```

#### 2.5% of them have made contact

```python
# b_gekauft_gesamt is the actual target variable since we are more interested on ones that got purchased.

gekauft_gesamt = (sum(cust_lead['b_gekauft_gesamt'])/len(cust_lead['b_gekauft_gesamt'].index))*100
gekauft_gesamt
```

#### 1.5% of them are purchased. This indicates the presence of class imbalance

```python
# Plotting the imbalance Analysis:
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize = (6,4))
plt.title('Imbalance Analysis',  fontsize=20)
chart = sns.countplot(data=cust_lead, x='b_gekauft_gesamt', palette='muted')
plt.xlabel('Purchased', fontsize=14)
plt.ylabel('count', fontsize=14)
```

#### This confirms that this is an imbalanced dataset
