```python
# Analysing columns with Object datatype and cleaning data

## Removed the row with invalid values and assigned the column a valid data type 
cust_lead = cust_lead[cust_lead.q_OpeningHours != 'ject']
cust_lead["q_OpeningHours"] = cust_lead.q_OpeningHours.astype(float)
```

```python
# Opening Days

print("Opening Days: " + str(sorted(cust_lead['q_OpeningDays'].unique())))
```

```python
# Because it has fixed values, creating a categorical variable from the field
new_col = cust_lead["q_OpeningDays"].astype("category")
cust_lead.insert(loc=12, column='q_OpeningDays_cat', value=new_col)
cust_lead.info()
```

```python
# Analysing dependent variables

kontakt_gewesen = (sum(cust_lead['b_in_kontakt_gewesen'])/len(cust_lead['b_in_kontakt_gewesen'].index))*100
kontakt_gewesen
```
